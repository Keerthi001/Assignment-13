## Import Sodium's YOLO or Solo

```
import sodium.yolo as solo
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
yolo = solo.YoloOpenCV()
```

## Download the images

```
import urllib.request
print('Beginning file download with urllib2...')
BASE_URL = 'https://github.com/Enamulla577/EVAI/tree/main/Session13/Images/'

file = 'mobile.jpeg'
urllib.request.urlretrieve(BASE_URL + file + '?raw=true', file)
```

```
img = yolo.run_yolo_on_image('mobile.jpeg')
plt.figure(figsize=(20, 15))
plt.imshow(img)
plt.axis('off')
plt.show()
```
